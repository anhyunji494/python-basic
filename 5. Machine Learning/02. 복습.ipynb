#### 머신러닝 복습
- 머신러닝이란? 
  : 지능을 가진 기계가 스스로 학습을 하여 패턴, 구조를 파악하고 새로운 데이터가 들어왔을 때 정답을 예측해내는 과정 

- 머신러닝의 종류 세가지 
  1) 지도학습 : 데이터에 대한 정답(Label)이 주어진 상태에서 학습
   - 분류 : 정답 데이터가 범주형
           => 클래스(선택할 수 있는 정답 데이터)
           
           - 이진분류 : 클래스 2개 
           - 다중분류 : 클래스 3개 이상
   - 회귀 : 정답 데이터가 연속형(수치형)
           
           
           - 정답의 미묘한 차이가 크게 중요하지 않음
  2) 비지도학습 : 데이터에 대한 정답이 없는 상태에서 학습
  3) 강화학습 : 보상을 주어 보상을 얻는 방향으로 학습

- 머신러닝의 7과정
  1) 문제 정의 
  2) 데이터 수집
  3) 데이터 전처리
  4) EDA (탐색적데이터분석) => 시각화, 기술통계확인
  5) 모델선택 (모델객체생성), 하이퍼파라미터 조정
  6) 모델학습, 예측
      - model.fit(훈련용문제, 훈련용정답)           
           - X : 문제, 특성, 피처, 속성, 입력, 독립변수
           - y : 정답, label, target, 종속변수
      - pre = model.predict(X_test)
  7) 모델 평가
   
      - accuracy_score(y_test, pre) => 정확도 (0~1)

#### 모델
- KNN (K-Nearest Neighbors / 최근접이웃모델)
    - 유유상종
    - k개만큼의 가장 가까운 데이터셋을 확인 
        => 새로운 데이터셋을 예측(분류:다수결/회귀:평균)
    - 하이퍼파라미터 : 이웃의 수 (n_neighbors)
        - 이웃의 수가 작을수록 과대적합 확률 상승 
        - 이웃의 수가 많이 설정될수록 과소적합 확률 상승
- Decision Tree (의사결정나무)
  
  
    - 예, 아니오로 결정을 하면서 분류를 진행 => 반복학습 
    - 의사 결정 방향 : 불순도가 낮아지는 방향 
    - 지니 불순도(Gini Impurity)
      - 해당 범주 안에 서로 다른 데이터가 얼마나 섞여 있는지
      - 결정 트리 모델의 노드 분할 기준 
      - 각 질문들이 얼마나 좋은 질문인지 수치로 파악
      - 0~0.5 사이 값을 범위로 가짐
      - 불순도가 0에 가까울수록 잘 분류된 것 (좋은 질문)
      - 불순도가 0.5라면 데이터가 5:5 비율로 섞여서 분류된 것 (좋지 않은 질문)
    - 하이퍼파라미터 : 사전 가지치기 (tree 모델은 과적합 확률이 높음)
      
      
      - max_depth : 트리의 최대 깊이 조절 
      - max_leaf_node : 리프 노드의 최대 개수 조절 
      - min_samples_split : 노드를 분할하기 위한 최소 샘플 수 
      - min_samples_leaf : 리프 노드가 가져야 할 최소 샘플 수
