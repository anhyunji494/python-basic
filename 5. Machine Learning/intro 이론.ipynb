##### 개요
인공지능>머신러닝>딥러닝
- 인공지능(Artificial Intelligence): 인간의 지적 능력을 컴퓨터를 통해 구현하는 기술
- 머신러닝(Machine Learning): 컴퓨터가 데이터를 통해 스스로 학습하여 예측이나 판단을 제공하는 기술
- 딥러닝(Deep Learning): 인공신경망(뉴런) 알고리즘을 활용하는 머신러닝 기술

#### 기계학습 (Machine Learning)
- 데이터를 기반으로 학습을 시켜서 모델을 만들고 예측/판단하게 만드는 기법
- 실세계의 복잡한 데이터로부터 규칙화 패턴을 발견하여 미래를 예측하는 기술






##### 인공지능의 정도(종류)



##### 머신러닝 종류

* 지도학습(Supervised Learning): 데이터+답 => 패턴 => 데이터를 넣으면 답을 예측 
* 비지도학습(Unsupervised Learning): 데이터 => 패턴/차원 축소
* 강화학습 (Reinforcement Learning): 정답-보상 / 오답-벌점 => 게임에서 자주 활용

#### 지도학습 (Supervised Learning)
- 데이터에 대한 Label(명시적인 답)이 주어진 상태에서 컴퓨터를 학습시키는 방법
- 예측하려는 값이 어떠한 값인지에 따라 분류(Classification)와 회귀(Regression)로 나뉨




> case 1) 분류: 혈액형, MBTI 
각 혈액형 : 클래스(범주, 명목)로 분류 
> case 2) 회귀: 연속적인 값을 분류 

x: 특성, 독립변수, 컬럼(부정확)
y: 레이블, 라벨, 타겟, 종속변수, 결과값 

##### 비지도학습(Unsuspervised Learning)
- 데이터에 대한 Label(명시적인 답)이 없는 상태에서 컴퓨터를 학습시키는 방법
- 데이터의 숨겨진 특징, 구조, 패턴을 파악하는데 사용
- 비슷한 특성의 데이터끼리 묵는 <b>클러스터링(Clustering), 차원 축소(Dimensionality Reduction)</b> 등이 있음

##### 강화학습(Reinforcement Learning)
- 지도학습과 비슷하지만, 완전한 답(Label)을 제공하지 않는 특징이 있음
- 더 많은 보상을 얻을 수 있는 방향으로 행동을 학습
- 주로 게임이나 로봇을 학습시키는데 많이 사용
